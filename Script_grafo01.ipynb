```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r'https://raw.githubusercontent.com/Jorge2018/DataScience2022-2/main/GHG-emissions-by-life-cycle-stage-OurWorldinData-upload.csv')
x1 = df['Food product']
y1 = df['Farm']
plt.figure(dpi=190)
plt.plot(x1, y1, label = "$y_1 -> Tab1$" )
plt.xticks(rotation=85,fontsize=6)
plt.grid()
plt.show()
```
